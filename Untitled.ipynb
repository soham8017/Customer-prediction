```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans
customer_data= pd.read_csv(r"C:\Users\soham\Music\project MALL\Mall_Customers.csv")
```

```python
#choosing the annualincome and spending
x=customer_data.iloc[:,[3,4]].values
```

```python
print(x)
```

```python
#choose correct number of clusters
# wcss=within cluster sum of sqr
wcss=[]
for i in range(1,11):
    kmeans=KMeans(n_clusters=i,init='k-means++',random_state=42)
    kmeans.fit(x)  
    wcss.append(kmeans.inertia_)
```

```python
sns.set()
plt.plot(range(1,11),wcss)
plt.title('point graph')
plt.xlabel('number of cluster')
plt.ylabel('wcss')
plt.show()
```

```python
#optimum number of cluster=5 as per the graph
kmeans=KMeans(n_clusters=5,init='k-means++',random_state=0)

#return a label for each data point based on their cluster
Y=kmeans.fit_predict(x)
print(Y)
```

```python
import matplotlib.pyplot as plt


# Assuming kmeans is already fitted and X, Y are defined
plt.figure(figsize=(8,8))

# Plotting the clusters with different colors
plt.scatter(x[Y==0,0], x[Y==0,1], s=50, c='green', label='cluster 1')
plt.scatter(x[Y==1,0], x[Y==1,1], s=50, c='red', label='cluster 2')
plt.scatter(x[Y==2,0], x[Y==2,1], s=50, c='blue', label='cluster 3')
plt.scatter(x[Y==3,0], x[Y==3,1], s=50, c='yellow', label='cluster 4')
plt.scatter(x[Y==4,0], x[Y==4,1], s=50, c='violet', label='cluster 5')

# Plotting the centroids
plt.scatter(kmeans.cluster_centers_[:,0], kmeans.cluster_centers_[:,1], s=100, c='cyan', label='centroids')

# Adding labels and title
plt.title('Customer Group')
plt.xlabel('Annual Income')
plt.ylabel('Spending Score')

# Display the legend and the plot
plt.legend()
plt.show()
```

```python
from sklearn.metrics import precision_score, recall_score, f1_score, confusion_matrix,accuracy_score
from sklearn.preprocessing import LabelEncoder
import seaborn as sns
import matplotlib.pyplot as plt

# Label Encoding for Gender (as Gender is a categorical variable)
le = LabelEncoder()
customer_data['Gender'] = le.fit_transform(customer_data['Gender'])

# Assuming you have already applied KMeans and have 'predicted_labels' from kmeans.labels_
predicted_labels = kmeans.labels_  # Clustering results
true_labels = customer_data['Gender']  # True Gender labels

# Calculate Precision, Recall, and F1 Score
precision = precision_score(true_labels, predicted_labels, average='macro', zero_division=1)
recall = recall_score(true_labels, predicted_labels, average='macro', zero_division=1)
f1 = f1_score(true_labels, predicted_labels, average='macro', zero_division=1)

# Print Precision, Recall, and F1 Score
print(f"Precision: {precision:.4f}")
print(f"Recall: {recall:.4f}")
print(f"F1 Score: {f1:.4f}")

# Confusion Matrix
conf_matrix = confusion_matrix(true_labels, predicted_labels)

# Plot Confusion Matrix
plt.figure(figsize=(8,6))
sns.heatmap(conf_matrix, annot=True, fmt="d", cmap="Blues", xticklabels=le.classes_, yticklabels=le.classes_)
plt.title('Confusion Matrix')
plt.xlabel('Predicted Labels')
plt.ylabel('True Labels')
plt.show()

```

```python
new_data = pd.DataFrame({
    'Annual Income (k$)': [70, 120, 55],  # Example annual income for new customers
    'Spending Score (1-100)': [60, 30, 85]  # Example spending scores for new customers
})

# Assuming you have already trained a KMeans model, use the model to predict clusters for the new data
predicted_new_labels = kmeans.predict(new_data)

# For visualization, let's combine the new data's true labels (if available) and predicted labels
# For example, let's say true labels for the new data (Gender) are:
new_true_labels = np.array([1, 0, 1])  # Example true labels (0 = Male, 1 = Female)

# Calculate Precision, Recall, and F1 Score for new data (if true labels are available)
precision_new = precision_score(new_true_labels, predicted_new_labels, average='macro', zero_division=1)
recall_new = recall_score(new_true_labels, predicted_new_labels, average='macro', zero_division=1)
f1_new = f1_score(new_true_labels, predicted_new_labels, average='macro', zero_division=1)
accuracy = accuracy_score(true_labels, predicted_labels)

# Print the accuracy
print(f"Accuracy of KMeans Segmentation: {accuracy:.4f}")


# Print Precision, Recall, and F1 Score for new data
print(f"Precision (New Data): {precision_new:.4f}")
print(f"Recall (New Data): {recall_new:.4f}")
print(f"F1 Score (New Data): {f1_new:.4f}")

# Confusion Matrix for new data
conf_matrix_new = confusion_matrix(new_true_labels, predicted_new_labels)

# Plot Confusion Matrix for new data
plt.figure(figsize=(8,6))
sns.heatmap(conf_matrix_new, annot=True, fmt="d", cmap="Blues", xticklabels=le.classes_, yticklabels=le.classes_)
plt.title('Confusion Matrix (New Data)')
plt.xlabel('Predicted Labels')
plt.ylabel('True Labels')
plt.show()

# ---------------------------------------------
# Step 3: Combine Original and New Data Metrics (Optional)
# Combine the original and new data results if desired
combined_true_labels = np.concatenate((true_labels, new_true_labels))
combined_predicted_labels = np.concatenate((predicted_labels, predicted_new_labels))

# Calculate Precision, Recall, and F1 Score for the combined data
precision_combined = precision_score(combined_true_labels, combined_predicted_labels, average='macro', zero_division=1)
recall_combined = recall_score(combined_true_labels, combined_predicted_labels, average='macro', zero_division=1)
f1_combined = f1_score(combined_true_labels, combined_predicted_labels, average='macro', zero_division=1)

# Print Precision, Recall, and F1 Score for the combined data
print(f"Precision (Combined Data): {precision_combined:.4f}")
print(f"Recall (Combined Data): {recall_combined:.4f}")
print(f"F1 Score (Combined Data): {f1_combined:.4f}")

# Confusion Matrix for combined data
conf_matrix_combined = confusion_matrix(combined_true_labels, combined_predicted_labels)

# Plot Confusion Matrix for combined data
plt.figure(figsize=(8,6))
sns.heatmap(conf_matrix_combined, annot=True, fmt="d", cmap="Blues", xticklabels=le.classes_, yticklabels=le.classes_)
plt.title('Confusion Matrix (Combined Data)')
plt.xlabel('Predicted Labels')
plt.ylabel('True Labels')
plt.show()
```
